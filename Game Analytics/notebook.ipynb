```python
import pandas as pd
import matplotlib.pyplot as plt
from test_analytics import count_unique_players, count_unique_events, participant_statistics, event_statistics, timestamp_statistics, scene_name_count
from test_analytics import turns_completed_statistics, time_taken_statistics, total_turns_completed_statistics, games_completed_statistics, time_taken_in_game_statistics, time_spent_in_game_statistics

# Path to the CSV file
csv_file_path = './data/2024_04_18totc.csv'
```

```python
# Count unique players from "participantID"

unique_player_count = count_unique_players(csv_file_path)

print("Unique player count:", unique_player_count)
```

```python
# Calculate statistics for "participantId"

mean_participant, median_participant, mode_participant = participant_statistics(csv_file_path)

print("Mean participant ID:", mean_participant)
print("Median participant ID:", median_participant)
print("Mode participant ID:", mode_participant)
```

```python
# Count unique events from "eventName"

unique_event_count = count_unique_events(csv_file_path)

print("Unique event count:", unique_event_count)
```

```python
# Calculate statistics for "eventName"

mean_event, median_event, mode_event = event_statistics(csv_file_path)

print("Mean event count:", mean_event)
print("Median event count:", median_event)
print("Mode event count:", mode_event)
```

```python
# Calculate mean, median, and mode timestamps

mean_time, median_time, mode_time = timestamp_statistics(csv_file_path)

print("Mean timestamp:", mean_time)
print("Median timestamp:", median_time)
print("Mode timestamp:", mode_time)
```

```python
# Count occurrences of each "sceneName"

scene_counts = scene_name_count(csv_file_path)

print("Scene counts:\n", scene_counts)
```

```python
# Plot a bar chart for scene names

scene_counts = scene_name_count(csv_file_path)
scene_counts.plot(kind='bar', figsize=(12, 6))

plt.title('Occurrences of Each Scene Name')
plt.xlabel('Scene Name')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Plot a time series of events over time
data = pd.read_csv(csv_file_path)
data['timestamp'] = pd.to_datetime(data['timestamp'])
data.set_index('timestamp', inplace=True)

# Resample to daily counts
daily_counts = data.resample('D').size()

# Plot the time series
daily_counts.plot(figsize=(12, 6))
plt.title('Number of Events Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Events')
plt.show()
```

```python
# Calculate summary statistics for "turnsCompleted"

mean_turns, median_turns, mode_turns, min_turns, max_turns = turns_completed_statistics(csv_file_path)

print("Mean turns completed:", mean_turns)
print("Median turns completed:", median_turns)
print("Mode turns completed:", mode_turns)
print("Min turns completed:", min_turns)
print("Max turns completed:", max_turns)
```

```python
# Plot a histogram of "turnsCompleted"

data = pd.read_csv(csv_file_path)
data['turnsCompleted'].dropna().plot(kind='hist', bins=30, figsize=(12, 6))
plt.title('Distribution of Turns Completed')
plt.xlabel('Turns Completed')
plt.ylabel('Frequency')
plt.show()
```
